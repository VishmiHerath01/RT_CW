```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import os

# Path to your saved results
results_dir = "/content/drive/MyDrive/RT-MyCopy/CNNbasedMedicalSegmentation/results"

# Get all PNG files
result_files = sorted(glob.glob(os.path.join(results_dir, "*.png")))

# Display each result with its metrics
for img_path in result_files:
    # Load and display the image
    img = mpimg.imread(img_path)
    plt.figure(figsize=(15, 5))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Result from {os.path.basename(img_path)}')
    plt.show()

# Load and display metrics if you saved them
print("\nModel Performance Summary:")
print("==========================")
print("The visualizations show:")
print("1. Left: Original FLAIR image")
print("2. Middle: Ground truth segmentation")
print("3. Right: Model's prediction")
print("\nTo use these results in the future:")
print("1. Models are saved in: /content/drive/MyDrive/RT-MyCopy/CNNbasedMedicalSegmentation/models")
print("2. Visualizations are saved in: /content/drive/MyDrive/RT-MyCopy/CNNbasedMedicalSegmentation/results")
```
